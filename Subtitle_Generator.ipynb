```python
import whisper
from moviepy.editor import VideoFileClip
import os

def extract_audio(video_path, audio_path="audio.wav"):
    """
    Extracts audio from a video file and saves it as a WAV file.
    """
    video = VideoFileClip(video_path)
    video.audio.write_audiofile(audio_path)
    return audio_path

def transcribe_audio_to_srt(audio_path, model_type="medium"):
    """
    Transcribes audio to SRT format using Whisper and saves the subtitle file.
    """
    # Load the Whisper model
    model = whisper.load_model(model_type)
    # Transcribe audio
    result = model.transcribe(audio_path)

    # Generate SRT content
    srt_content = []
    for i, segment in enumerate(result["segments"]):
        start = segment["start"]
        end = segment["end"]
        text = segment["text"]

        # Format timestamps
        start_str = format_timestamp(start)
        end_str = format_timestamp(end)

        # Append each subtitle line in SRT format
        srt_content.append(f"{i + 1}\n{start_str} --> {end_str}\n{text}\n")

    # Save to an SRT file
    srt_path = audio_path.replace(".wav", ".srt")
    with open(srt_path, "w") as f:
        f.writelines(srt_content)

    print(f"Subtitles saved to {srt_path}")

def format_timestamp(seconds):
    """
    Formats time in seconds to SRT timestamp format (HH:MM:SS,MS).
    """
    hours = int(seconds // 3600)
    minutes = int((seconds % 3600) // 60)
    seconds = int(seconds % 60)
    milliseconds = int((seconds % 1) * 1000)
    return f"{hours:02}:{minutes:02}:{seconds:02},{milliseconds:03}"

# Main script
video_path = "/content/drive/MyDrive/SoccerNet/video_2023/england_epl/2014-2015/2015-05-17 - 18-00 Manchester United 1 - 1 Arsenal/1_224p.mkv"
audio_path = extract_audio(video_path)
transcribe_audio_to_srt(audio_path)
```

```python
def clean_srt_file(input_srt_path, output_srt_path):
    """
    Cleans an SRT file by removing leading spaces from each line and ensuring each entry has a blank line after it.

    Parameters:
        input_srt_path (str): Path to the input SRT file.
        output_srt_path (str): Path to save the cleaned SRT file.
    """
    with open(input_srt_path, 'r', encoding='utf-8') as infile, open(output_srt_path, 'w', encoding='utf-8') as outfile:
        for line in infile:
            # Remove leading/trailing spaces and write the cleaned line
            cleaned_line = line.strip()

            # Write the cleaned line to output file
            outfile.write(cleaned_line + '\n')

            # Add an extra newline after each subtitle entry (text block ends when next line is empty)
            if cleaned_line == "":
                outfile.write('\n' + '\n')


    print(f"Cleaned SRT file saved as {output_srt_path}")

# Example usage
input_srt_path = "audio.srt"
output_srt_path = "clean.srt"
clean_srt_file(input_srt_path, output_srt_path)
```

```python
import subprocess

def write_subtitles_to_video(video_path, srt_path, output_path):
    """
    Embeds subtitles into a video file using ffmpeg.

    Parameters:
        video_path (str): Path to the input video file.
        srt_path (str): Path to the SRT subtitle file.
        output_path (str): Path where the output video with subtitles will be saved.
    """
    # Command to use ffmpeg to burn subtitles into the video
    command = [
        'ffmpeg',
        '-i', video_path,        # Input video
        '-vf', f"subtitles={srt_path}",  # Subtitle filter
        '-c:a', 'copy',          # Copy audio without re-encoding
        output_path              # Output video file
    ]

    # Run the command
    try:
        subprocess.run(command, check=True)
        print(f"Video with subtitles saved to {output_path}")
    except subprocess.CalledProcessError as e:
        print(f"Error embedding subtitles: {e}")

# Example usage
video_path = "/content/England vs Spain World Cup Highlights.mp4"
srt_path = "audio.srt"
output_path = f"{video_path.split('.mp4')[0]}_subtitled.mp4"
write_subtitles_to_video(video_path, srt_path, output_path)
```

```python
import torch
from transformers import VideoMAEModel, AutoTokenizer, VideoMAEProcessor
import av  # for reading videos
import numpy as np

# Load the model, tokenizer, and processor
model_name = "model_name_or_path"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = VideoMAEModel.from_pretrained(model_name)
image_processor = VideoMAEProcessor.from_pretrained(model_name)

device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)

# Freeze encoder layers
for param in model.encoder.parameters():
    param.requires_grad = False

# Define loss and optimizer
criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.AdamW(filter(lambda p: p.requires_grad, model.parameters()), lr=5e-5)

# Load video and extract frames
video_path = "clip.mp4"
container = av.open(video_path)

seg_len = container.streams.video[0].frames
clip_len = model.config.encoder.num_frames
indices = set(np.linspace(0, seg_len, num=clip_len, endpoint=False).astype(np.int64))
frames = []
container.seek(0)

for i, frame in enumerate(container.decode(video=0)):
    if i in indices:
        frames.append(frame.to_ndarray(format="rgb24"))

# Process frames
pixel_values = image_processor(frames, return_tensors="pt").pixel_values.to(device)

# Dummy captions for training
train_captions = ["A man and a woman are dancing on a stage in front of a mirror."]
inputs = tokenizer(train_captions, return_tensors="pt", padding="max_length", max_length=20, truncation=True)
labels = inputs.input_ids.to(device)

# Training loop
num_epochs = 5
for epoch in range(num_epochs):
    model.train()

    # Forward pass
    outputs = model(pixel_values, labels=labels)
    loss = criterion(outputs.logits.view(-1, model.config.vocab_size), labels.view(-1))

    # Backward pass and optimization
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    print(f"Epoch [{epoch + 1}/{num_epochs}], Loss: {loss.item():.4f}")

# Generate caption
model.eval()
gen_kwargs = {
    "min_length": 10,
    "max_length": 20,
    "num_beams": 8,
}
tokens = model.generate(pixel_values, **gen_kwargs)
caption = tokenizer.batch_decode(tokens, skip_special_tokens=True)[0]
print(caption)
```

```python
import av
import numpy as np
import torch
from transformers import AutoImageProcessor, AutoTokenizer, VisionEncoderDecoderModel

# Set device
device = "cuda" if torch.cuda.is_available() else "cpu"

# Load pretrained processor, tokenizer, and model
image_processor = AutoImageProcessor.from_pretrained("MCG-NJU/videomae-base")
tokenizer = AutoTokenizer.from_pretrained("gpt2")
model = VisionEncoderDecoderModel.from_pretrained("Neleac/timesformer-gpt2-video-captioning").to(device)

# Freeze encoder parameters
for param in model.encoder.parameters():
    param.requires_grad = False

# Define optimizer and loss
criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.AdamW(filter(lambda p: p.requires_grad, model.parameters()), lr=5e-5)

# Load and process video
video_path = "clip.mp4"
container = av.open(video_path)
seg_len = container.streams.video[0].frames
clip_len = model.config.encoder.num_frames
indices = set(np.linspace(0, seg_len, num=clip_len, endpoint=False).astype(np.int64))
frames = []

container.seek(0)
for i, frame in enumerate(container.decode(video=0)):
    if i in indices:
        frames.append(frame.to_ndarray(format="rgb24"))

# Prepare pixel values
pixel_values = image_processor(frames, return_tensors="pt").pixel_values.to(device)

# Example training captions
train_captions = ["A man and a woman are dancing on a stage in front of a mirror."]

# Assign a padding token if not defined
if tokenizer.pad_token is None:
    tokenizer.pad_token = tokenizer.eos_token

# Tokenize captions with padding and truncation
inputs = tokenizer(train_captions, return_tensors="pt", padding="max_length", max_length=20, truncation=True)
labels = inputs.input_ids.to(device)

# Mask padding tokens in labels
labels[labels == tokenizer.pad_token_id] = -100


# Training loop
num_epochs = 5
for epoch in range(num_epochs):
    model.train()
    optimizer.zero_grad()

    # Forward pass
    outputs = model(pixel_values, labels=labels)
    loss = outputs.loss

    # Backward pass and optimization
    loss.backward()
    optimizer.step()

    print(f"Epoch [{epoch + 1}/{num_epochs}], Loss: {loss.item():.4f}")

# Evaluate and generate captions
model.eval()
gen_kwargs = {
    "min_length": 10,
    "max_length": 20,
    "num_beams": 8,
}
with torch.no_grad():
    tokens = model.generate(pixel_values, **gen_kwargs)
    caption = tokenizer.batch_decode(tokens, skip_special_tokens=True)[0]
    print(f"Generated Caption: {caption}")
```
